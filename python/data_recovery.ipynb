```python
import csv
import re
```

```python
TARGET_FILE_NAME = "../gitignore/asymptote/medium_HF_2023_Aug_19_14_24_04_.csv"
LOG_FILE_NAME = "../gitignore/asymptote/medium_HF_2023_Aug_19_14_24_04_log.txt"
OUTER_LINE = re.compile('Generated \d+ line program\. ([^W]+) Wrote out to "([^"]+)"\.')
INNER_LINE = re.compile('Measured p-value: ([\d.e-]+)')
```

```python
with open(TARGET_FILE_NAME) as f:
    text = f.read()
headers = text.split(",")
headers
```

```python
with open(LOG_FILE_NAME) as f:
    lines = [l for l in f]
lines[0]
```

```python
data = []
for l in lines:
    m = OUTER_LINE.match(l)
    if m:
        measurements = m.group(1)
        name = m.group(2)
        mms = INNER_LINE.findall(measurements)
        data.append([name] + [float(mm) for mm in mms])
data
```

```python
s = "\n".join(','.join(str(d) for d in dd) for dd in data)
s
```

```python
with open(TARGET_FILE_NAME, 'a') as f:
    f.write(s)
```
